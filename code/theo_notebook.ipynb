```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

# Read the data
data = pd.read_csv('data/chocolate.csv')
```

```python
# percentage of chocolate
data['cocoa_percent'] = data['cocoa_percent'].str.replace('%', '').astype(float)
px.box(data, x='cocoa_percent', points="all").show()
```

```python
# chocolate rating
px.box(data, x='rating', points="all").show()
```
