```python
import numpy as np
import pandas as pd
import plotly.express as px

# Read the data
data = pd.read_csv('data/chocolate.csv')

data.head()
```

```python
# vérifier qu'il ne manque pas de données
print(data.isnull().sum())

# vérifier qu'il n'y a pas de doublons
print(data.duplicated().sum())

```

```python
# Afficher le nombre de chocolats par pays d'origine
fig = px.histogram(data, x='country_of_bean_origin', title='Nombre de chocolats par pays d\'origine').update_xaxes(categoryorder='total descending')
fig.show()
fig=px.pie(data, names='country_of_bean_origin', title='Pays d\'origine des fèves de cacao').update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
tab=data['country_of_bean_origin'].value_counts()

# Origine des fèves de cacao par pays
fig = px.scatter_geo(tab, locations=tab.index, hover_name=tab.index, locationmode='country names', size=tab.values, title='Origine des fèves de cacao par pays')
fig.show()
```
